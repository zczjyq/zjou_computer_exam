```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 读取数据
data = pd.read_excel('./recourse/kaoshibaoExcel.xlsx')
data = data[:1]
```

```python
import pandas as pd
# 获取判断题
soup = BeautifulSoup(open('./recourse/1.html')) 
div_all_pds = soup.select('div.pc-x.pt-2.pl-4')
# 列数
num_columns = len(data.columns)

for i in range(len(div_all_pds)):
    try:
        text = div_all_pds[i].p.text
        new_row = pd.DataFrame([[text, "判断题", "T", "F"] + [None] * (num_columns - 4)], columns=data.columns)
        data = pd.concat([data, new_row], ignore_index=True)
    except:
        text = div_all_pds[i].li.text
        new_row = pd.DataFrame([[text, "判断题", "T", "F"] + [None] * (num_columns - 4)], columns=data.columns)
        data = pd.concat([data, new_row], ignore_index=True)
        print(f"KeyError: at index {i}")
        print(div_all_pds[i])

# print(data)
```

```python
def select_item(new_row_text, subject, j):
    try:
        new_row_text.append(subject[j].p.text)
    except:
        try:
            new_row_text.append(subject[j].li.text)
        except:
            try:
                new_row_text.append(subject[j].code.text)
            except:
                print("三级错误")
    return new_row_text
```

```python
# 读取单选题数据
soup = BeautifulSoup(open('./recourse/单选题.html')) 
div_all_xuanzes = soup.select('div.pc-v.pc-gap-12')

for i in range(len(div_all_xuanzes)):
    subject = div_all_xuanzes[i].select("div.markdownBlock_tErSz")
    new_row_text = []
    # print(subject)
    for j in range(len(subject)):
        
        new_row_text = select_item(new_row_text, subject, j)
        if j == 0:
            new_row_text.append("单选题")
    new_row = pd.DataFrame([new_row_text + [None] * (num_columns - len(new_row_text))], columns=data.columns)
    data = pd.concat([data, new_row], ignore_index=True)
    # print(len(new_row_text))
        # print(j)
    # new_row = pd.DataFrame([[text, "判断题", "T", "F"] + [None] * (num_columns - 4)], columns=data.columns)
    
```
