```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
soup = BeautifulSoup(open('./recourse/填空题.html')) 
```

```python
all_tiankong = soup.select("div.pc-x.pt-2.pl-4")
```

```python
markdown_content = ""
index = 1
for tiankong in all_tiankong:
    markdown_content += f"{index}. " + tiankong.p.text.replace("1 分", "_____").replace("2 分  ", "_____").replace("3 分  ", "_____") + "\n"
    index += 1
with open(f'数据库填空题.md', 'w', encoding='utf-8') as f:
    f.write(markdown_content)
```
